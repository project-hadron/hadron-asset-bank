```python
gen = np.random.default_rng()
```

```python
sb = SyntheticBuilder.from_memory()
```

------------------

```python
sb.set_persist_uri('./source/hadron_synth_origin.pq')
```

```python
df = sb.tools.model_synthetic_data_types(10, extended=True)
```

```python
df_syn = df.copy().iloc[:,np.r_[0:9,16]]

# add rows
df_rows = sb.tools.model_synthetic_data_types(2, extended=True, seed=31).iloc[:,np.r_[0:9,16]]
df_rows['unique'] = df['unique'] + 100
df_ext = pd.concat([df_rows, df_syn], axis=0)
df_ext = df_ext.sample(frac=1).reset_index(drop=True)
df_ext = df_ext.iloc[:,gen.choice(range(df_ext.columns.size), df_ext.columns.size, replace=False)]
```

```python
sb.save_persist_canonical(df_ext)
```

```python
sb.set_persist_uri('./source/hadron_synth_other.pq')
```

```python
df_syn = df.copy().iloc[:,np.r_[0:6,10,16]]
```

```python
df_syn['num'] = sb.tools.correlate_values(df_syn, header='num', choice=5, jitter=2)
df_syn['int'] = sb.tools.correlate_values(df_syn, header='int', choice=2, jitter=2, precision=0)
df_syn['cat'] = sb.tools.correlate_categories(df_syn, header='cat', correlations=['PENDING'], actions={0: 'WAITING'}, default_action=sb.tools.action2dict(method='@header', header='cat'))
```

```python
# add rows
df_rows = sb.tools.model_synthetic_data_types(5, extended=True, seed=31).iloc[:,np.r_[0:6,10,16]]
df_rows['unique'] = df['unique'] + 200
df_syn = pd.concat([df_rows, df_syn], axis=0)
# shuffle
df_syn = df_syn.sample(frac=1).reset_index(drop=True)
df_syn = df_syn.iloc[:,gen.choice(range(df_syn.columns.size), df_syn.columns.size, replace=False)]
```

```python
sb.save_persist_canonical(df_syn)
```
